# Sensex

```python
sensex = yf.Ticker("^BSESN")
print(sensex.info)

sensex_hist = sensex.history(period="max")
sensex_hist
```

```python
sensex_hist_mean = sensex_hist.rolling(window = 28).mean()
sensex_hist_mean
```

```python
rolling_mean = sensex_hist['Close'].rolling(window=120).mean()

# Plot the original 'Close' values
plt.figure(figsize=(10, 6))
plt.plot(sensex_hist['Close'], label='Close')

# Plot the rolling mean
plt.plot(rolling_mean, label='Rolling Mean (window=120)', color='red')

# Set labels and title
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Sensex Close and Rolling Mean')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Assuming sensex_hist is your DataFrame
sensex_hist['Date'] = pd.to_datetime(sensex_hist.index)

# Fill or interpolate missing values in 'Close' column
sensex_hist['Close'].fillna(method='ffill', inplace=True)  # Forward fill

# Calculate log returns
sensex_hist['Log_Returns'] = sensex_hist['Close'].pct_change().apply(lambda x: np.log(1 + x))

# Calculate the moving mean, moving skewness, moving kurtosis, and moving excess kurtosis for log returns
window_size = 60
sensex_hist['Moving_Mean'] = sensex_hist['Log_Returns'].rolling(window=window_size).mean()
sensex_hist['Moving_Skewness'] = sensex_hist['Log_Returns'].rolling(window=window_size).apply(lambda x: skew(x, nan_policy='omit'))
sensex_hist['Moving_Kurtosis'] = sensex_hist['Log_Returns'].rolling(window=window_size).apply(lambda x: kurtosis(x, nan_policy='omit'))
sensex_hist['Moving_Excess_Kurtosis'] = sensex_hist['Moving_Kurtosis'] - 3  # Excess kurtosis is calculated by subtracting 3

# Plotting
plt.figure(figsize=(12, 10))

plt.subplot(4, 1, 1)
plt.plot(sensex_hist['Date'], sensex_hist['Moving_Mean'], label='Moving Mean (Log Returns)', color='blue')
plt.title('Moving Mean vs Date')
plt.xlabel('Date')
plt.ylabel('Moving Mean')
plt.legend()

plt.subplot(4, 1, 2)
plt.plot(sensex_hist['Date'], sensex_hist['Moving_Skewness'], label='Moving Skewness (Log Returns)', color='green')
plt.title('Moving Skewness vs Date')
plt.xlabel('Date')
plt.ylabel('Moving Skewness')
plt.legend()

plt.subplot(4, 1, 3)
plt.plot(sensex_hist['Date'], sensex_hist['Moving_Kurtosis'], label='Moving Kurtosis (Log Returns)', color='red')
plt.title('Moving Kurtosis vs Date')
plt.xlabel('Date')
plt.ylabel('Moving Kurtosis')
plt.legend()

plt.subplot(4, 1, 4)
plt.plot(sensex_hist['Date'], sensex_hist['Moving_Excess_Kurtosis'], label='Moving Excess Kurtosis (Log Returns)', color='orange')
plt.title('Moving Excess Kurtosis vs Date')
plt.xlabel('Date')
plt.ylabel('Moving Excess Kurtosis')
plt.legend()

plt.tight_layout()
plt.show()
```
